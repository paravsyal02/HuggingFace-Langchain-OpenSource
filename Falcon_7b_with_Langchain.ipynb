```python
from langchain import HuggingFacePipeline
from transformers import AutoTokenizer, pipeline
import torch

model ="tiiuae/falcon-7b-instruct"

tokenizer = AutoTokenizer.from_pretrained(model)

pipeline= pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.bfloat16,
    trust_remote_code=True,
    device_map="auto",
    max_length=200,
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id
)
```

```python
llm = HuggingFacePipeline(pipeline=pipeline, model_kwargs={"temperature":0})
```

```python
from langchain import PromptTemplate, LLMChain
```

```python
template = """
You are an inteligent chatbot. Help the following question with brilliant answers

Question: {question}
Answer:
"""
prompt = PromptTemplate(template=template, input_variables=["question"])
```

```python
llm_chain = LLMChain(prompt=prompt, llm=llm)
```

```python
question = "Give me a code for adding 2 numbers"

print(llm_chain.run(question))
```

```python
question = "Explain Artificial Intelligence using poem"

print(llm_chain.run(question))
```

```python
question = "Explain Artificial Intelligence using math"

print(llm_chain.run(question))
```
