```python
from llama_index.core import VectorStoreIndex, SimpleDirectoryReader, Settings, PromptTemplate
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from llama_index.llms.groq import Groq
```

```python
!mkdir data
```

```python
documents = SimpleDirectoryReader("/content/data").load_data()
```

```python
documents
```

```python
embed_model = HuggingFaceEmbedding(model_name="sentence-transformers/all-MiniLM-L6-v2")
Settings.embed_model = embed_model
```

```python
template = """You are a Q&A assistant. Your goal is to answer questions as accurately as possible based on the instructions and context provided.

Context: {context}
Question: {query}

Answer:"""
prompt_template = PromptTemplate(template=template)
```

```python
groq_llm = Groq(
    model="mixtral-8x7b-32768",
    api_key="************************************",
    prompt_template=prompt_template
)
Settings.llm = groq_llm
```

```python
index = VectorStoreIndex.from_documents(documents)
```

```python
## Q/a operation
query_engine = index.as_query_engine()
```

```python
response = query_engine.query("What is inheritance tell me in detail?")
print(response)
```
